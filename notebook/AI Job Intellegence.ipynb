# Job Market Overview

```python
df['job_count'] = 1
```

```python
df['jobs_in_country'] = df.groupby('country')['job_count'].transform('sum')
```

```python
df[['country','jobs_in_country']].drop_duplicates().sort_values('jobs_in_country',ascending = False)
```

```python
# Insights : AI jobs varies narrowly ranging from 7350 to 7602 across countries. Canada reports the highest number of job roles,while Japan the lowest.
```

```python
df['jobs_in_industry'] = df.groupby('industry')['job_count'].transform('sum')
```

```python
df[['industry','jobs_in_industry']].drop_duplicates().sort_values('jobs_in_industry',ascending = False)
```

```python
# Insights : Gaming & Education industry leads the ai job market hiring while automative & Healthcare industry are trailing
```

```python
df['jobs_in_role'] = df.groupby('job_role')['job_count'].transform('sum')
```

```python
df[['job_role','jobs_in_role']].drop_duplicates().sort_values('jobs_in_role',ascending = False)
```

```python
# Insights : All roles relatively have uniform range of job count varies from 11063 to 11412,indicating demand is evenly distributed across AI job
# market. NLP Engineers reports the highest number of job count, while Data Analysts the lowest.
```

```python
df['jobs_in_company'] = df.groupby('company_size')['job_count'].transform('sum')
```

```python
df[['company_size','jobs_in_company']].drop_duplicates().sort_values('jobs_in_company',ascending = False)
```

```python
# Insights : Job counts are relatively uniform across company sizes(17757-18328).Medium sized companies reports the highest number of jobs , while Large 
# companies the lowest.
```

```python
df.to_csv('job_market_overview.csv',index = False)
```

#  Career Growth Analysis

```python
# How AI careers evolve over time and which paths are most rewarding
```

```python
salary_growth = df.groupby('experience_level')['salary'].mean().round(2).sort_values(ascending = False)
salary_growth
```

```python
# Insights  :  Salary increases significanly with experience , longest jump is observed between Mid & Senior level
```

```python
spec_growth = df.groupby(['ai_specialization','experience_level'])['salary'].mean().round(2).reset_index(name = 'mean_salary')
```

```python
spec_growth['rank'] = spec_growth.groupby('experience_level')['mean_salary'].rank(method = 'dense',ascending = False)
```

```python
spec_growth.sort_values(['experience_level','rank'],ascending = True)
```

```python
# Insights : At Entry & Medium level Analytics focused specialization offer the highest career growth .
# However at Senior levels, core AI specialization such as Generative AI,Mlops show the highest salaries, suggesting that deeper technical 
# expertise become more valuable with experience.
```

# Job Stability & Risk Analysis

```python
df['layoff_risk_in_industry'] = df.groupby('industry')['layoff_risk'].transform('mean').round(2)
```

```python
layoff_risk_in_industry = df.groupby('industry')['layoff_risk'].mean().round(2).sort_values(ascending = False)
```

```python
layoff_risk_in_industry
```

```python
# Insights : All industries exhibit the same layoff risk(0.18), indicating that layoff risk are driven by external factors like
# employee skill relevance, market condition and economic trends rather than industry specific dynamics.
```

```python
df['automation_risk_in_role'] = df.groupby('job_role')['automation_risk'].transform('mean').round(2)
```

```python
automation_risk_in_role = df.groupby('job_role')['automation_risk'].mean().round(2).sort_values(ascending = False)
automation_risk_in_role
```

```python
# Insights : Automation risk is uniform accross all roles (~50), suggesting that automation impact is evenly distributes accross
# AI & data related roles.However,Computer Vision Engineer shows the highest risk(50.64) while AI Engineers have the lowest(50.09)
```

```python
df[['salary','layoff_risk']].corr()
```

```python
# Insigts : A moderare negative correlation between salary and layoff risk shows that higher paid roles tend to have slightly lower risk of
# layoffs.
```

```python
df['job_security_in_role'] = df.groupby('job_role')['job_security_score'].transform('mean').round(2)
```

```python
job_security = df.groupby('job_role')['job_security_score'].mean().round(2).sort_values(ascending = False)
job_security
```

```python
# Insights : All job roles have uniform job security ranging from 75.78 to 75.45. However Data Scientists have the highest job security(~75.78)
# ,while Data Analyst have the lowest job security score (~75.45),indication minimal variations between roles.
```

```python
df['risk_in_company'] = df.groupby('company_size')['layoff_risk'].transform('mean').round(2)
```

```python
risk_in_company = df.groupby('company_size')['layoff_risk'].mean().round(2).sort_values(ascending = False)
risk_in_company
```

```python
# Insights : Layoff risk is uniform accross all company sizes(~0.18),indicating that layoffs are driven by external factors like 
# employee skills and market conditions rather than company size.
```

```python
df['risk_in_exp'] = df.groupby('experience_level')['layoff_risk'].transform('mean').round(2)
```

```python
risk_in_experience = df.groupby('experience_level')['layoff_risk'].mean().round(2).sort_values(ascending = False)
risk_in_experience
```

```python
# Insights : Layoff risk decreases sharply from entry(0.38) to mid level(0.18) and stabilizes at senior levels(0.08),indicating that 
# higher experience, technical expertise and decision making responsibility reduces layoff risks.
```
